```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re
from sklearn import linear_model
```

```python
df12 = pd.read_csv('/Users/brucehu/Desktop/Dec.csv')
```

```python
print(df12)
```

```python
plt.style.use('ggplot')
```

```python
ax1 = df12[df12['price'] >= 1200].plot(kind='scatter', x='dis', y='score', color='red', alpha=0.5, figsize=(10, 7))
df12[df12['price'] < 1200].plot(kind='scatter', x='dis', y='score', color='green', alpha=0.5, figsize=(10 ,7), ax=ax1)
plt.legend(labels=['1200+', 'low'])
plt.title('Relationship between dis and score', size=24)
plt.xlabel('dis', size=18)
plt.ylabel('score', size=18)
```

```python
df_high = df12[df12['price']>=1200]
df_low = df12[df12['price']<1200]

high_fit = np.polyfit(df_high.dis, df_high.score, 1)
low_fit = np.polyfit(df_low.dis, df_low.score, 1)

ax1 = df_high.plot(kind='scatter', x='dis', y='score', color='red', alpha=0.5, figsize=(10, 7))
df_low.plot(kind='scatter', x='dis', y='score', color='green', alpha=0.5, figsize=(10, 7), ax=ax1)

plt.plot(df_high.dis, high_fit[0] * df_high.dis + high_fit[1], color='darkred', linewidth=2)
plt.plot(df_low.dis, low_fit[0] * df_low.dis + low_fit[1], color='lightblue', linewidth=2)
```
